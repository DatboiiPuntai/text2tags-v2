```python
filtered_df = df[
    (df["tag_count_general"] >= 15) &
    (df["tag_count_general"] <= 80) &
    (df["is_banned"] == False) &
    (df["is_deleted"] == False) &
    (df["rating"] == 's') &
    (df['created_at'].dt.year > 2015)
]
filtered_df
```

```python
tag_df = pd.read_json(r"data/danbooru2021/tags000000000000.json", lines=True)
tag_df
```

```python
filtered_tags = tag_df[(tag_df["category"] == 0) & (tag_df["post_count"] >= 600)].sort_values(
    "post_count", ascending=False
).reset_index(drop=True)
filtered_tags
```

```python
with open(r'data/dictionaries/tag_list.txt', 'w') as f:
    f.write('\n'.join(filtered_tags.name.to_list()))
```

```python
filtered_df["filtered_tag_string"] = [
    [tag for tag in tags if tag in filtered_tags.name]
    for tags in filtered_df["tag_string_general"].str.split()
]

filtered_df["filtered_tag_string"] = filtered_df["filtered_tag_string"].apply(lambda x: " ".join(x))
```

```python
from collections import defaultdict

def count_tags(df):
    tag_counts = defaultdict(lambda: {'post_count': 0})
    for _, row in df.iterrows():
        tags = row['tag_string_general'].split()
        for tag in tags:
            tag_counts[tag]['post_count'] += 1
    return tag_counts


def create_df(tag_counts):
    tag_df = pd.DataFrame.from_dict(tag_counts, orient='index')
    tag_df.index.name = 'name'
    tag_df = tag_df.sort_values('post_count', ascending=False)
    tag_df = tag_df.reset_index()
    return tag_df
```

```python
tag_counts = count_tags(filtered_df)
post_tag_df = create_df(tag_counts)
post_tag_df
```

```python
open('bruh.json', 'w').write(post_tag_df.to_json(lines=True, orient='records'))
```
